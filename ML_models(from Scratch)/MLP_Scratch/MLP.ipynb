# Multilayer Perceptron (From Scratch)

This code represents a simple neural network from scratch by only utilizing numpy.

```python
# Import necessary libraries 
import numpy as np
import matplotlib.pyplot as plt
```

## Dataset

```python
# Generating spiral function shaped dataset
def create_data():
    N = 100  # number of points per class
    D = 2  # dimensionality
    K = 3  # number of classes
    X = np.zeros((N * K, D))  # data matrix (each row = single example)
    y = np.zeros(N * K, dtype="uint8")  # class labels
    for j in range(K):
        ix = range(N * j, N * (j + 1))
        r = np.linspace(0.0, 1, N)  # radius
        t = np.linspace(j * 4, (j + 1) * 4, N) + np.random.randn(N) * 0.2  # theta
        X[ix] = np.c_[r * np.sin(t), r * np.cos(t)]
        y[ix] = j
    return X, y
```

Dataset contains 3 different classes which were visualized in th figure below. This dataset requires a nonlinear classifer to successfully distinguish the hidden three classes.

```python
X, y = create_data()

#Visualizing the dataset
plt.figure(figsize=(12,9))
plt.scatter(X[:, 0], X[:, 1], c=y, s=40)
plt.show()
```

```python
# Creating the neural network structure
class NN_Layer:
    def __init__(self, n_inputs, n_neurons):
        # Creating a random weight with specified matrix size
        # n_inputs = number of inputs
        # n_neurons = number of neurons inside the layer
        # randn: gaussian distribution rounded around 0
        self.weights = 0.10 * np.random.randn(
            n_inputs, n_neurons
        )  # Multiplying by 0*10 to achieve values less than 1

        self.biases = np.zeros((1, n_neurons))

    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases


layer1 = NN_Layer(2,5)
```

```python
layer1.forward(X[:5, :])
layer1.out
```
